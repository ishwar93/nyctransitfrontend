```python
# validate_isochrones.py
# Streams the file — never loads more than one feature at a time
import ijson, sys

REQUIRED_KEYS = {"h3", "walkBand", "walkabilityMult", "coverageCount",
                 "stopTypes", "hasSubway", "hasBus", "hasCitibike"}
VALID_BANDS   = {5, 10, 15}

path = "isochrones.geojson"
count = 0
bad_schema = []
bad_bands  = []
walk_bands_seen = set()
sample_props = None

with open(path, "rb") as f:
    for feat in ijson.items(f, "features.item"):
        count += 1
        props = feat.get("properties", {})

        # Capture first feature for display
        if sample_props is None:
            sample_props = props

        # Check all required keys present
        missing = REQUIRED_KEYS - props.keys()
        if missing:
            bad_schema.append((count, missing))

        # Check walkBand is 5, 10, or 15
        wb = props.get("walkBand")
        walk_bands_seen.add(wb)
        if wb not in VALID_BANDS:
            bad_bands.append((count, wb))

        # Stop after first 3 errors of each type to avoid log spam
        if len(bad_schema) >= 3 and len(bad_bands) >= 3:
            print(f"Stopping early at feature {count} — errors found")
            break

        if count % 500_000 == 0:
            print(f"  ... scanned {count:,} features")

print(f"\nTotal features scanned: {count:,}")
print(f"walkBand values found:  {sorted(walk_bands_seen)}")
print(f"Schema errors:          {len(bad_schema)}")
print(f"Bad walkBand values:    {len(bad_bands)}")
print(f"\nSample properties (feature 1):")
for k, v in sample_props.items():
    print(f"  {k}: {v!r}")
```
